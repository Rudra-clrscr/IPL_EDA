# IMPORTING IMPORTANT LIBRARIES :-

```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# DOCUMENT LOADING :-

```python
df=pd.read_csv("D:/Downloads/matches.csv")
```

# READING THE DATA :-

```python
df.info()
df.describe()
```

# CHECKING FOR MISSING VALUES :-

```python
df.isnull()
df.isnull().sum()
```

# HANDLING MISSING VALUES :-

```python

df.drop('umpire3',axis=1,inplace=True)
df.dropna(inplace=True)


```

```python
df.isnull()
df.isnull().sum()
```

# EDA TECHNIQUES :-

# 1. Total matches won by each team (Bar Chart)

```python

plt.figure(figsize=(12, 6))
sns.countplot(y='winner', data=df, order=df['winner'].value_counts().index, hue='winner',palette='viridis',legend=False)
plt.title('Total Matches Won by Each Team')
plt.xlabel('Number of Matches Won')
plt.ylabel('Team')
plt.tight_layout()
plt.savefig('Total_matches_won.png')
plt.show()
```

# 2. Matches hosted by each venue (Bar Chart)

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='venue', data=df, order=df['venue'].value_counts().index, hue='venue',palette='plasma',legend=False)
plt.title('Total Matches Hosted by Each Venue')
plt.xlabel('Number of Matches Hosted')
plt.ylabel('Venue')
plt.tight_layout()
plt.savefig('Matches_hosted_by.png')
plt.show()
```

# 3. Distribution of wins by runs (Histogram)

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['win_by_runs'],bins=15,kde=True, color='purple')
plt.title('Distribution of Wins by Runs')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('Distribution_of_wins_by_runs.png')
plt.show()
```

# 4. Distribution of wins by wickets (Histogram)

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['win_by_wickets'], bins=15, kde=True, color='teal')
plt.title('Distribution of Wins by Wickets')
plt.xlabel('Wickets')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('Distribution_of_wins_by_wickets.png')
plt.show()
```

# 5. IPL Trophies Won by Each Team (2008-2017) (Barchart)

```python
df = df.sort_values(by=['season', 'id'])
final_matches = df.drop_duplicates(subset='season', keep='last')
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='winner',data=final_matches,order=final_matches['winner'].value_counts().index, color='skyblue')
plt.title('IPL Trophies Won by Each Team (2008-2017)')
plt.xlabel('Team')
plt.ylabel('Number of Trophies')
plt.tight_layout()
plt.savefig('IPL_trophies_won_by.png')
plt.show()
```

# 6. Top 10 Players in the IPL (Bar Chart)

```python
plt.figure(figsize=(12,6))
sns.countplot(y='player_of_match',data=df,order=df['player_of_match'].value_counts().head(10).index,color='orange')
plt.title('Top 10 Player in the IPL')
plt.xlabel('Runs')
plt.ylabel('Players')
plt.tight_layout()
plt.savefig('Top_10_players.png')
plt.show()
```

# Some special systaxes used in this code :-
#### . plt.figure() -> used to set size of the graph
#### . plt.title() -> used to set title of the graph
#### . plt.show() -> used to display the graph
#### . plt.xlabel() -> used to set name of x-axis
#### . plt.ylabel() -> used to set name of y-axis
#### . plt.tight_layout() -> used so that the bars are fit to the graph
#### . sns.countplot() -> used to create a bar chart on the number of counts of the elements in the column
#### . sns.histplot() -> used to create a histogram for a column
